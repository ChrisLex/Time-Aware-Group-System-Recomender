```python
##### Paso 3: Aplicación de estrategias para crear ratings grupales bloque 5

print('Empieza paso 3:')
print("Aplicación de estrategias para crear ratings grupales")

#### size=2
groups = pd.read_csv("Tesis/item-KNN/1.Baseline/grupos/group_2.csv", sep=',' )

# solo hay dos usuarios por eso se crea la U1 y U2
members = pd.DataFrame(groups.MembersID.str.split('\t',2).tolist(), 
                     columns = ['U1','U2']
                    )
# Haciendo el merge para tener todos los usuarios
groupid_users = pd.concat([groups['GroupID'], members], axis=1)

# Para tener en forma horizontal 
groups1 = groupid_users[['GroupID', 'U1']]
groups2 = groupid_users[['GroupID', 'U2']]
groups2 = groups2.rename(columns={'U2': 'U1'}, inplace=False)

# Haciendo el merge groups1 y groups2
groups_1_2 = pd.concat([groups1, groups2])
# cambiando el tipo de dato
groups_1_2['U1'] = groups_1_2['U1'].astype('float').astype('int64')

# Leyendo los ratings que están ya llenados, para jusntarle con los grupos creados y sus usuarios.
ratings_data = pd.read_csv( "RatingsPredictionsFinales005-5.csv", sep=',')

# Haciendo el inner join de groups con ratings de las movies
groups_movies = pd.merge(groups_1_2, ratings_data, left_on='U1', right_on='userId')
groups_movies = groups_movies[['GroupID', 'userId', 'mmovieId', 'rating']]

# guardando en un csv el group_movies_size2
groups_movies.to_csv('Tesis/item-KNN/1.Baseline/agregaciones/group_movies_size2_5.csv', index=False)

# ahora hagamos el pivot
group_movie_rating = groups_movies.pivot_table(index=['GroupID','userId'], columns='mmovieId', values='rating')

print('n=2')
print("Additive strategy")

group_movie_AVG = groups_movies.pivot_table(index=['GroupID'], columns='mmovieId', values='rating', 
                                            aggfunc=np.sum)
group_movie_AVG = group_movie_AVG.reset_index()

# despivoteando la tabla de average staretgy
group_movie_AVG_unpivot = group_movie_AVG.melt(id_vars=['GroupID'], var_name='mmovieId', value_name='RatingsGroups')

##### normalizando
# calculo del max_group_rating
# es el valor maximo de rating grupal que puede ser obtenido por un item 
max_group_rating = group_movie_AVG_unpivot.groupby('mmovieId')['RatingsGroups'].max()
max_group_rating = max_group_rating.reset_index()

# Rename the columns 
max_group_rating = max_group_rating.rename(columns = {'RatingsGroups': 'max_group_rating'}, inplace = False)

# el max_user_rating
# es el valor maximo de rating individual que un usuario puede expresar para un item.
max_user_rating = groups_movies.groupby('mmovieId')['rating'].max()
max_user_rating =  max_user_rating.reset_index()

# Rename the columns 
max_user_rating = max_user_rating.rename(columns = {'rating': 'max_user_rating'}, inplace = False)

group_movie_ADD_desnorm = pd.merge(pd.merge(group_movie_AVG_unpivot, max_group_rating, on='mmovieId'),
                                   max_user_rating, on='mmovieId')

group_movie_ADD_desnorm['NewRatingsGroups'] = (group_movie_ADD_desnorm['RatingsGroups'] * group_movie_ADD_desnorm['max_user_rating'])/(group_movie_ADD_desnorm['max_group_rating'])

group_movie_ADD_desnorm = group_movie_ADD_desnorm[['GroupID', 'mmovieId', 'NewRatingsGroups']]

# guardando en un csv el group_movie_AVG_unpivot size1
group_movie_ADD_desnorm.to_csv('Tesis/item-KNN/1.Baseline/agregaciones/group_movies_size2_ADD_5.csv', 
                               index=False)

#Multiplicative Strategy
group_movie_MULT = groups_movies.pivot_table(index=['GroupID'], columns='mmovieId', values='rating', aggfunc=np.prod)
group_movie_MULT = group_movie_MULT.reset_index()

# despivoteando la tabla de average staretgy
group_movie_MULT_unpivot = group_movie_MULT.melt(id_vars=['GroupID'], var_name='mmovieId', value_name='RatingsGroups')

# calculo del max_group_rating
# es el valor maximo de rating grupal que puede ser obtenido por un item 
max_group_rating = group_movie_MULT_unpivot.groupby('mmovieId')['RatingsGroups'].max()
max_group_rating = max_group_rating.reset_index()

# Rename the columns 
max_group_rating = max_group_rating.rename(columns = {'RatingsGroups': 'max_group_rating'}, inplace = False)

# el max_user_rating
# es el valor maximo de rating individual que un usuario puede expresar para un item.
max_user_rating = groups_movies.groupby('mmovieId')['rating'].max()
max_user_rating =  max_user_rating.reset_index()

# Rename the columns 
max_user_rating = max_user_rating.rename(columns = {'rating': 'max_user_rating'}, inplace = False)

group_movie_MULT_desnorm = pd.merge(pd.merge(group_movie_MULT_unpivot, max_group_rating, on='mmovieId'),
                                   max_user_rating, on='mmovieId')

group_movie_MULT_desnorm['NewRatingsGroups'] = (group_movie_MULT_desnorm['RatingsGroups'] * group_movie_MULT_desnorm['max_user_rating'])/(group_movie_MULT_desnorm['max_group_rating'])

group_movie_MULT_desnorm = group_movie_MULT_desnorm[['GroupID', 'mmovieId', 'NewRatingsGroups']]

# guardando en un csv el group_movie_AVG_unpivot size1
group_movie_MULT_desnorm.to_csv('Tesis/item-KNN/1.Baseline/agregaciones/group_movies_size2_MULT_5.csv', 
                                index=False)

##### Least Misery Strategy

group_movie_LEAST = groups_movies.pivot_table(index=['GroupID'], columns='mmovieId', values='rating', aggfunc=np.min)
group_movie_LEAST = group_movie_LEAST.reset_index()

# despivoteando la tabla de average staretgy
group_movie_LEAST_unpivot = group_movie_LEAST.melt(id_vars=['GroupID'], var_name='mmovieId', value_name='RatingsGroups')

# guardando en un csv el group_movie_AVG_unpivot size1
group_movie_LEAST_unpivot.to_csv('Tesis/item-KNN/1.Baseline/agregaciones/group_movies_size2_LMS_5.csv',
                                 index=False)

#### Most Pleasure Strategy
group_movie_MOST = groups_movies.pivot_table(index=['GroupID'], columns='mmovieId', values='rating', aggfunc=np.max)
group_movie_MOST = group_movie_MOST.reset_index()

# despivoteando la tabla de average staretgy
group_movie_MOST_unpivot = group_movie_MOST.melt(id_vars=['GroupID'], var_name='mmovieId', value_name='RatingsGroups')


# guardando en un csv el group_movie_AVG_unpivot size1
group_movie_MOST_unpivot.to_csv('Tesis/item-KNN/1.Baseline/agregaciones/group_movies_size2_MPS_5.csv', 
                               index=False)
```
