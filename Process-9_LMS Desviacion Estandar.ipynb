```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
  
df = pd.read_csv("Tesis/item-KNN/1.Baseline/agregaciones/group_movies_size2_LMS.csv")
df8 = pd.read_csv("Tesis/item-KNN/1.Baseline/agregaciones/group_movies_size8_LMS.csv")
dataframe1 = df['RatingsGroups'].std(axis= 0)
print("The Standard Deviation (size 2) is: \n")
print(dataframe1)
dataframe8 = df8['RatingsGroups'].std(axis= 0)
print("The Standard Deviation (size 8) is: \n")
print(dataframe8)

df['RatingsGroups8'] = df8['RatingsGroups']

df.boxplot(column =['RatingsGroups','RatingsGroups8'], grid = False)


plt.title("LMS T1")
plt.suptitle("")
plt.ylabel("RatingsGroups")
plt.show()


```
